```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function a(n) = floor((n-1)/2)
def a_n(n):
    return np.floor((n - 1) / 2)

# Define the smoothed sum S(N) for a range of large N values
def smoothed_sum(N, max_n=1000):
    n_values = np.arange(1, max_n + 1)
    a_values = (-1) ** a_n(n_values)  # Compute (-1)^a(n)
    exp_values = np.exp(-n_values / N)  # Apply the regulator e^(-n/N)
    return np.sum(a_values * exp_values)

# Define a range of large N values
N_values = np.linspace(10, 1000, 50)
S_values = np.array([smoothed_sum(N) for N in N_values])

# Plot the asymptotic behavior
plt.figure(figsize=(8, 5))
plt.plot(N_values, S_values, label=r'$S(N) = \sum_n (-1)^{a(n)} e^{-n/N}$', color='b')
plt.axhline(y=1, color='r', linestyle='--', label="Expected Value = 1")
plt.xlabel("N")
plt.ylabel("Smoothed Sum S(N)")
plt.title("Asymptotic Behavior of Smoothed Sum with e^(-n/N) Regulator")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Fujikawa regulator and general periodic series
# Define the function a(n) = floor((n-1)/2)
def a_n(n):
    return np.floor((n - 1) / 2)

# Define the smoothed sum S(N) for Fujikawa reg e^{-n^2/N^2}
def smoothed_sum_gaussian(N, max_n=5000):
    n_values = np.arange(1, max_n + 1)
    a_values = (-1) ** a_n(n_values)  # Compute (-1)^a(n)
    exp_values = np.exp(-n_values**2 / N**2)  # Apply reg e^(-n^2/N^2)
    return np.sum(a_values * exp_values)

# Define range of large N values
N_values = np.linspace(10, 20000, 50)
S_values_gaussian = np.array([smoothed_sum_gaussian(N) for N in N_values])

# Plot the asymptotic behaviour
plt.figure(figsize=(8, 5))
plt.plot(N_values, S_values_gaussian, label=r'$S(N) = \sum_n (-1)^{a(n)} e^{-n^2/N^2}$', color='g')
plt.xlabel("N")
plt.ylabel("Smoothed sum S(N)")
plt.title("Asymptotic behaviour of periodic sum with Fujikawa regulator e^(-n^2/N^2)")
plt.legend()
plt.grid()
plt.show()
```
